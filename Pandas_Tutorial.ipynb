Pandas is one of the most important libraries of Python.

Pandas has data structures for easy data analysis. The most used of these are Series and DataFrame data structures. Series data structure is one dimensional, that is, it consists of a column. DataFrame data structure is two-dimensional, i.e. it consists of rows and columns.

# Create DataFrame from dicts

```python
data={"name":["Bill","Tom","Tim","John","Alex","Vanessa","Kate"],      
      "score":[90,80,85,75,95,60,65],      
      "sport":["Wrestling","Football","Skiing","Swimming","Tennis",
               "Karete","Surfing"],      
      "sex":["M","M","M","M","F","F","F"]}
```

```python
df=pd.DataFrame(data)
df
```

```python
df=pd.DataFrame(data,columns=["name","sport","sex","score"])
df
```

```python
df.head(3)
```

```python
df.tail(3)
```

```python
df=pd.DataFrame(data,columns=["name", "sport", "gender", "score", "age"],
                index=["one","two","three","four","five","six","seven"])
df
```

```python
df[["sport","score"]]
```

#  Useful Functions

```python
data=pd.read_csv("DataSets/ca_san_diego_2019_02_25.csv")
data
```

```python
data=data.dropna()
data
```

## Describing Table

```python
data.describe()
```

```python
#accumulating cols from index with iloc
data.iloc[0]
```

```python
## finding unique values
string=data["subject_race"].unique()
string
```

```python
new_data={t:i for i,t in enumerate(string) }
print(new_data)
new_coloum=[]
for i in data["subject_race"].values:
    new_coloum.append(new_data[i])
data["numerical_subject_race"]=new_coloum
```

```python
data
```

```python
data=data.drop_duplicates()
```

```python
data
```

```python
## Computing Correlation 
data.corr()
```

```python
## Finding Min or Max 
print("Minimum Age",data['subject_age'].min())
print("Maximum Age",data['subject_age'].max())
## Finding maximum or minimum index
min_index=data['subject_age'].idxmin()
max_index=data['subject_age'].idxmax()
print("Sex of minimum age",data['subject_sex'][min_index])
print("Sex of maximum age",data['subject_sex'][max_index])
print("Reasons for minimum age",data['reason_for_stop'][min_index])
print("Reasons maximum age",data['reason_for_stop'][max_index])
```

```python
## Calculating mean and median  
print("Average of Age",data['subject_age'].mean())
print("Median of Age",data['subject_age'].median())
```

```python
## Counting the number of each feature 
print(data['subject_age'].count())
print(data['subject_age'].value_counts())
```
